```python
import json

def convert(obj):
    l = []
    try:
        # Parse the string as JSON
        parsed_obj = json.loads(obj)
        # Extract the 'name' key for the first 3 dictionaries
        l = [i['name'] for i in parsed_obj[:3]]  # Get only the first 3 names
    except (json.JSONDecodeError, KeyError, TypeError):
        # Handle invalid JSON strings or missing 'name' keys gracefully
        print(obj)
    return l
```

```python
import json

def convert(obj):
    l = []
    try:
        # Parse the string as JSON
        parsed_obj = json.loads(obj)
        # Extract the 'name' key for the first 3 dictionaries
        l = [i['name'] for i in parsed_obj[:2]]  # Get only the first 3 names
    except (json.JSONDecodeError, KeyError, TypeError) as e:
        # Handle invalid JSON strings or other issues gracefully
        print(f"Skipping invalid entry: {obj} -> Error: {e}")
    return l
```

```python
data['genres']=data['genres'].apply(lambda x:[i.replace(" ","") for i in x])
data['keywords']=data['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
data['overview']=data['overview'].apply(lambda x:[i.replace(" ","") for i in x])
data['cast']=data['cast'].apply(lambda x:[i.replace(" ","") for i in x])
data['crew']=data['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def str(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
data['tags']=data['tags'].apply(str)
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

```python
tf=TfidfVectorizer()
```

```python
vector=tf.fit_transform(data['tags']).toarray()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
similarity[0]
```

```python
def recomended(movie):
    index=data[data['title']==movie].index[0]
    distance=similarity[index]
    movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movie_list:
        print(data.iloc[i[0]].title)    
```

```python
recomended('Aliens')
```
